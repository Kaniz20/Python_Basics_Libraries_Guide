```python
#import seaborn & pandas library
import seaborn as sns
import pandas as pd
```

```python
#load data
df=sns.load_dataset("penguins")
```

```python
df.head()
```

```python
df.isnull().sum()
```

```python
#class level analysis
df["species"].unique()
```

#Numerical Plot

```python
#Distribution plot(one feature)
sns.histplot(data=df,x="flipper_length_mm",kde=True,hue="island",legend="auto")
```

```python
#Joinplot(For two feature)
sns.jointplot(data=df, x="bill_length_mm", y="bill_depth_mm", hue="species")
```

```python
#pairplot
sns.pairplot(data=df)
#sns.pairplot(data=df, kind='reg')
```

```python
#pairplot
#sns.pairplot(data=df)
sns.pairplot(data=df, kind='reg')
```

#Categorical data

```python
# Countplot
sns.countplot(data=df,x="sex",palette="PuBu",width=0.5)
```

```python
#barplot(categorical & numerical both)
sns.barplot(df, x="flipper_length_mm", y="sex",  hue='island',orient="hx",width=0.5)
```

```python
#boxplot
sns.boxplot(data=df,y="flipper_length_mm", x='island', hue='species',palette="PuBu")
```

```python
#violineplot
sns.violinplot(data=df, x="flipper_length_mm", y="sex", legend='auto',width=0.6,palette="PuBu")
```

```python
#pairGrid
g = sns.PairGrid(df)
g.map(sns.scatterplot)
#g.map_diag(sns.histplot)
#g.map_offdiag(sns.kdeplot)
g.add_legend()
```
