```python
#import matplotlib library
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
```

```python
x = [10, 20, 30, 40, 50, 60]
y = [100, 200, 300, 400, 550, 260]
#plt.plot(x,y,color="red",linestyle="--",linewidth=1.6)
plt.plot(x,y,"b--*")
plt.xlabel("Price")
plt.ylabel("Amount[kg]")
plt.title("This is my matplotlib chart")
```

```python
x1 = [10, 20, 30, 40, 50, 60]
y1 = [100, 200, 300, 400, 500, 600]
y2 = [50, 60, 70, 80, 90, 100]
y3 = [200, 400, 600, 800, 1000, 1200]
y4 = [70, 100, 200, 150, 180, 133]
plt.plot(x1,y1,"r--*", label="Age")
plt.plot(x1,y2,"g--*", label="Name")
plt.plot(x1,y3,"b--*", label="Hobby")
plt.plot(x1,y4,"y-*",label="Class")
plt.grid()
plt.legend(loc="best",shadow=True,fontsize=10)
plt.xlabel("Price")
plt.ylabel("Amount[kg]")
plt.title("This is my matplotlib chart")
```

#Sub Plot

```python
plt.subplot(2,2,1)
plt.plot(x1,y1,"r--*", label="Age")
plt.grid()
plt.subplot(2,2,2)
plt.plot(x1,y2,"g--*", label="Name")
plt.grid()
plt.subplot(2,2,3)
plt.plot(x1,y3,"b--*", label="Hobby")
plt.grid()
plt.subplot(2,2,4)
plt.plot(x1,y4,"y-*",label="Class")
plt.grid()
plt.show()


```

#Bar Chart Create

```python
name = ['dhrubo', 'sabit', 'siam', 'imran', 'kobri', 'rokib']
roll = [191, 22, 133, 70, 77, 80]
number = [99, 80, 70, 71, 66, 50]
plt.bar(name,roll,label="roll",width=0.4,color="blue")
plt.bar(name,number,label="number",width=0.4)
plt.legend()
```

```python
name = ['dhrubo', 'sabit', 'siam', 'imran', 'kobri', 'rokib']
a=np.arange(len(name))
roll = [191, 22, 133, 70, 77, 80]
number = [99, 80, 70, 71, 66, 50]
plt.bar(a-0.2,roll,label="roll",width=0.4)
plt.bar(a+0.2,number,label="number",width=0.4)
plt.legend()
```

```python
name = ['dhrubo', 'sabit', 'siam', 'imran', 'kobri', 'rokib']
a=np.arange(len(name))
roll = [191, 22, 133, 70, 77, 80]
number = [99, 80, 70, 71, 66, 50]
plt.barh(a-0.2,roll,label="roll",height=0.4)
plt.barh(a+0.2,number,label="number",height=0.4)
plt.legend()
```

```python
#histtype='bar' ---by default type bar

number = [99, 80, 80, 70, 70, 88, 155, 166, 167, 71, 71, 266, 50, 55, 80, 58, 70, 255, 166, 80]
plt.hist(number, bins=[50, 80, 100, 150, 160], rwidth=0.95, histtype='bar', color='y')
```

```python
men = [99, 80, 80, 70, 70, 70, 155, 166, 167, 71, 71, 266, 50, 55, 80, 58]
women = [69, 70, 80, 70, 71, 72, 157, 126, 117, 81, 41, 256, 60, 59, 82]
plt.hist([men, women], bins=[50,80,100, 130, 150, 160, 200, 250, 300], color=['r', 'g'], label=['man', 'women'], rwidth=0.95, orientation='horizontal')
plt.legend()
```

```python

expense = [500, 1500, 200, 900]
exp_name = ['food', 'rent', 'bill', 'others']
colors = ['r', 'g', 'b', 'gold']
explode = [0.0, 0.0, 0.0, 0.1]
# plt.pie(expense, labels=exp_name, colors=colors, explode=explode, autopct="%0.2f%%")
plt.pie(expense, labels=exp_name, explode=explode, autopct="%.2f%%", colors=colors, counterclock=False)
plt.show()
```
